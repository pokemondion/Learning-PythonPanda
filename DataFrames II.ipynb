## Filter with More than One Condition (AND)

```python
df = pd.read_csv("employees.csv", parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
# extract males and marketing team

mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2].head()
```

## Filter with More than One Condition (OR)

```python
df = pd.read_csv("employees.csv", parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
# extract True senior management or before jan 1st, 1990

mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"

df[mask1 | mask2].head()
```

```python
#either a name of robert and in client services, or after the date June 1st,2016

mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"

mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3].head()
```

## The .isin() Method

```python
df = pd.read_csv("employees.csv", parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
# a little bit inefficient 
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3].head()
```

```python
# go thru every row for the team column and check if its in either of the values given in the list
# a lot more adaptable 
mask = df["Team"].isin(["Legal","Sales","Product","Marketing"])
df[mask].head()
```

## The .duplicated Method (Series)

```python
df = pd.read_csv("employees.csv", parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
# extract rows that are duplicates
# marks subsequent values as duplicates after the first one

# 'first', 'last', False

df[df["First Name"].duplicated()].head()
```

```python
df["First Name"].duplicated(keep = 'last').head()
```

```python
df[df["First Name"].duplicated(keep = False)].head() # marks true if it is duplicated more than once
                                              # keeps all duplicates and OG
```

```python
mask = ~df["First Name"].duplicated(keep = False) # get the unique values, values than only occur once in the first name column
df[mask].tail()
```

## the .drop_duplicates() Method (DataFrames)

```python
df = pd.read_csv("employees.csv", parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
len(df)
```

```python
# drop_duplicates on its own removes rows that are exactly the same in each column
# must customize the drop_duplicates() method

len(df.drop_duplicates())
```

```python
df.drop_duplicates(subset = ["First Name"], keep = "first").head() # keep the first one even if unique or duplicated in the df
```

```python
df.drop_duplicates(subset = ["First Name"], keep = False).tail() # keeps all non-duplicates
```

```python
# will get an empty df if there are no duplicates
df.drop_duplicates(subset = ["Team"], keep = False).head() # drop duplicates and keep nothing
```

```python
df.drop_duplicates(subset = ["First Name","Team"]).tail() # removes the duplciates of the same "First Name" AND "Team"
```
